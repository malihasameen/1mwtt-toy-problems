```python
# peakhell sound like pickle (Python Object Serialization)
import urllib.request
import pickle

html = urllib.request.urlopen("http://www.pythonchallenge.com/pc/def/banner.p")
html
```

```python
data = pickle.load(html)
data
```

```python
for line in data:
    print("".join(k * v for k,v in line))
```
