```python
import urllib.request
```

```python
html = urllib.request.urlopen("http://www.pythonchallenge.com/pc/def/ocr.html").read().decode()
html
```

```python
import re
```

```python
comments = re.findall("<!--(.*?)-->", html, re.DOTALL)
comments
```

```python
data = comments[1]
data
```

```python
count = {}
for c in data:
    count[c] = count.get(c,0) + 1
count
```
